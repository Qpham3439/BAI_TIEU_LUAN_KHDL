CELL 2 — Đọc dữ liệu
Giải thích:

Đọc dataset từ file CSV vào DataFrame.

```python
df = pd.read_csv(
    "../data/Hotel Reservations.csv"
)

df.head()
```

CELL 4 — Xem thông tin dữ liệu

CELL 5 — Làm sạch dữ liệu
Giải thích:

Xóa bản ghi trùng lặp và loại bỏ các cột không cần thiết.

```python
df = df.drop_duplicates()

df = df.drop(
    columns=[
        "Booking_ID",
        "booking_status"
    ]
)

df.head()
```

CELL 6 — Chọn thuộc tính
Giải thích:

Giữ lại các thuộc tính phản ánh hành vi khách hàng.

```python
features = [

"lead_time",

"avg_price_per_room",

"no_of_special_requests",

"repeated_guest"

]

X = df[features]

X.head()
```

CELL 7 — Chuẩn hóa dữ liệu
Giải thích:

Các thuộc tính có đơn vị khác nhau nên cần chuẩn hóa.

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(
    X
)
```

CELL 8 — Tìm số cụm tối ưu
Giải thích:

Thử nhiều giá trị K và đánh giá bằng Silhouette Score.

```python
for k in range(2,11):

    model = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=20
    )

    labels = model.fit_predict(
        X_scaled
    )

    score = silhouette_score(
        X_scaled,
        labels
    )

    print(
        f"K={k}: {score:.3f}"
    )
```

CELL 8 — Tìm số cụm tối ưu
